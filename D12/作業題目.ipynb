```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
df = pd.read_csv('boston.csv')
df = df.set_index('key')
plot = df.boxplot(figsize=(15,10))
[{item_name} for item_name in df.columns[(df.median()>300)&(df.median()<400)]]
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
plot = df.plot.scatter(x='NOX', y='DIS')
print('NOX 越大, DIS 越小')
```
